```python
from train_util import *

total = 21

fig = plt.figure(figsize=(12,2*total)) 

train_x_mat_files = sorted([f for f in os.listdir('train/1') if f.endswith('.mat')])
data_x_mat_files =  sorted([f for f in os.listdir('data') if f.endswith('.mat')])

for i, id  in enumerate(np.random.randint(400, 499, 6)):
    for k in range(total):
        plt.subplot(total + 1,6,i + k * 6 + 1)
        img = scipy.io.loadmat(os.path.join('out/'+str(k+1), train_x_mat_files[id]))['value']
        plt.imshow(img)

    plt.subplot(total + 1,6,i + 1 + 6 * total)
    img = scipy.io.loadmat(os.path.join('data', data_x_mat_files[id]))['value']
    plt.imshow(img)
    
plt.tight_layout()
plt.show()

fig.savefig('figures/export_fig.png', dpi=fig.dpi)
```
